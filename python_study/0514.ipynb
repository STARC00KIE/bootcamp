## 2. 그리기 함수 활용

### (1) plt.plot() 함수안에 올 수 있는 데이터
- list, ndarray, series, dataframe

````python
# list
plt.plot([1, 2, 3], [4, 7, 8])
plt.show()
````

````python
# ndarray
plt.plot(np.array([1, 2, 3]), np.array([4, 5, 6]))
plt.show()
````

````python
# Series
plt.plot(pd.Series([1, 2, 3]), pd.Series([4, 5, 6]))
plt.show()
````

````python
df =pd.DataFrame(np.random.randn(10, 4), columns=['A', 'B', 'C', 'D'],
                index=np.arange(0, 100, 10))
df
````

````python
plt.plot(df)
plt.show()
````

### (2) pandas의 내장 matplotlib
- pandas에 matplotlib이 내장되어 있어서
- Series.plot(), DataFrame.plot() 형태로도 호출할 수 있음

**주요 plot 옵션**
- figsize(그래프 크기), kind(그래프 종류), title(제목)
- xlable(x축 제목), ylabel(y축 제목), legend(범례)
- grid(격자), rot(눈금회전), subplots(복수 그래프)

````python
df.plot(kind='line', legend=True, title='df line plot')
plt.show()
````

````python
df.plot(kind='bar', legend=True, title='df line plot', grid=True)
plt.show()
````

### (4) x축, y축에 축명을 텍스트로 할당

````python
plt.plot([1, 2, 3], [4, 5, 6], color='green', linestyle='dashed', linewidth=3, marker='o', markersize=10)
plt.xlabel('x axis', size=20)
plt.ylabel('y axis', size=20)
plt.show()
````

### (5) x축, y축 틱값 표기

````python
plt.plot([1, 2, 3], [4, 5, 6], color='green', linestyle='dashed', linewidth=3, marker='o', markersize=10)
plt.xlabel('x axis', size=20)
plt.ylabel('y axis', size=20)

plt.xticks([1, 2, 3], rotation=90)
plt.yticks([4, 5, 6])
plt.show()
````

### (6) x축, y축값 제한

````python
plt.plot([1, 2, 3], [4, 5, 6], color='green', linestyle='dashed', linewidth=3, marker='o', markersize=10)
plt.xlabel('x axis', size=20)
plt.ylabel('y axis', size=20)

plt.xticks([1, 2, 3], rotation=30)
plt.yticks([4, 5, 6])

plt.xlim(0, 2)
plt.ylim(3, 5)
plt.show()
````

### (7) 범례 설정하기

````python
plt.plot([1, 2, 3], [4, 5, 6], label='test', color='green', linestyle='dashed', linewidth=3, marker='o', markersize=10)
plt.xlabel('x axis', size=20)
plt.ylabel('y axis', size=20)

plt.xticks([1, 2, 3], rotation=30)
plt.yticks([4, 5, 6])

plt.xlim(0, 2)
plt.ylim(3, 5)

plt.legend()
plt.show()
````

### (8) 여러 종류의 그래프를 하나의 axes에서 그리기

````python
x1 = np.arange(10)
y1 = x1 * 2

plt.plot(x1, y1, color='red', marker='+', label='line') # 선그래프
plt.bar(x1, y1, color='green', label='bar') # 막대그래프

plt.xlabel('x axis', size=20)
plt.ylabel('y axis', size=20)

plt.legend()
plt.show()
````

````python
# 위의 코드를 fig, axes가 하는일을 알수 있게 바꾼 코드 (동일한 결과)
x1 = np.arange(10)
y1 = x1 * 2

fig = plt.figure()
axes = plt.axes()

axes.plot(x1, y1, color='red', marker='+', label='line') # 선그래프
axes.bar(x1, y1, color='green', label='bar') # 막대그래프

axes.set_xlabel('x axis', size=20)
axes.set_ylabel('y axis', size=20)

axes.legend()
plt.show()
````
