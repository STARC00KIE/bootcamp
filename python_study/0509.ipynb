# Matplotlib이란?
- 다양한 데이터를 파이썬에서 시각화할 수 있게 해주는 라이브러리

# 관련 표현
- Figure (도화지): 전체 그래프 영역, 한 화면.

- Axes (축): 실제로 그래프를 그리는 작은 그래프 영역.

- Subplot (서브플롯): Figure를 격자로 나눈 작은 영역 = Axes가 들어가는 자리.

# **1. Matplotlib 실행**
### **도화지, 서브플롯 생성**
  * 도화지 생성
  ```
  plt.figure(figsize = (x,y))
  ```
  * 서브플롯 및 축 추가
  ```
  .add_subplot(x,y,z)
  ```
  * 도화지, 서브플롯 한번에 생성
  ```
  fig, axes = plt.subplots(2,3, figsize = (6,3))
  # 2,3 : 축의 행, 열 수
  # 6,3 : 도화지 크기 지정
  ```
  * 한번에 만들어진 서브플롯에 각 그래프 그리기
  ```
  axes[x,y].그래프(data)
  ```





### **그래프 그리기**
  * 히스토그램
  ```
  plt.hist(y, bins = 막대 수, color = 색상, alpha = 투명도)
  ```

  * 산점도
  ```
  plt.scatter(x, y)
  ```

  * 선 그래프
  ```
  plt.plot(x, y)   # '--' 등으로 선 종류 지정 가능
```


### **그래프 디테일 설정**
- .set_title('~~') : 그래프 제목 지정
- .set_xlabel('~~') : x축 이름 지정
- fig.subplots_adjust() : 도화지 여백 설정. 상하좌우 다 조정 가능


````python
## 간단한 그래프 그리기 ##

import numpy as np
data = np.arange(10)    # 시각화 할 데이터 생성

plt.plot(data)          # 데이터를 선 그래프로

plt.show()              # 그래프 시각화
````

````python
## 서브플롯 객체에서 플롯하기 ##

# 서브플롯 객체: 한 화면(도화지) 안에 여러 개의 작은 그래프(서브플롯)를 만들 수 있게 해주는 개별 그래프 영역
# 서브플롯 객체에서 플롯하기 = 개별 그래프 영역에서 시각화해보기


# 1. 새로운 Figure(도화지)를 생성하고, 크기를 가로 4인치 x 세로 3인치로 설정
fig = plt.figure(figsize = (4,3))


# 2. .add_subplot(x,y,z) : x행 y열의 서브플롯 중 z번째 위치에 축 추가
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(223)        # 이렇게 줄여 작성해도 인식됨


# 3. 각 축에 그래프 그리기
ax1.hist(np.random.randn(100), bins = 20, color = 'g', alpha = 0.3)     # 'g' : green
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
ax3.plot(np.random.randn(50).cumsum(), 'b--')                           # .cumsum() : 누적 합을 구하는 NumPy 함수
                                                                        # 'k--'

# 4. 모든 플롯을 화면에 표시
plt.show()


# 5. 열려 있는 모든 플롯 창 닫기 (리소스 정리)
plt.close('all')
````

````python
## 한꺼번에 여러 서브플롯 객체 생성하기 ##

fig, axes = plt.subplots(2,3, figsize = (6,3))
# 2행 3열의 서브플롯 생성
# 전체 도화지 크기는 가로 6인치 * 세로 3인치



# 참고 ↓

## 한꺼번에 만든 서브플롯에 각 그래프 넣기 ##

# axes[x,y].plot(data) - 위치 지정으로 바로 그래프 삽입 가능

axes[0,0].plot(data, 'g*')
axes[1,2].boxplot(data)

plt.show()
plt.close()
````

````python
## 색상, 마커, 선 스타일 연습 ##

# 도화지 생성
fig = plt.figure(figsize = (4,3))

# 선 그래프 그리기
plt.plot(np.random.randn(30).cumsum(), 'mo:')
#      'm' : 마젠타 색상
#      'o' : 점 모양을 원으로
#      ':' : 점선 스타일

plt.show()
plt.close()
````

````python
## 눈금, 라벨 ##

# 제목, 축 이름, 눈금, 눈금 이름 설정
fig = plt.figure(figsize = (5,2))
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())

# 눈금 설정
ticks = ax.set_xticks([0,250,500,750,1000])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation = 45, fontsize = 'small')

# 제목, 축 이름 설정
ax.set_title('그래프 제목')
ax.set_xlabel('x축 제목')

# 도화지 여백 설정
fig.subplots_adjust(bottom = 0.3)

plt.show()
plt.close('all')
````
