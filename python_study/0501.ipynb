```python
# 1. space 를 기준으로 word 를 구분하고
# 2. 3글자 이하는 제외하고
# 3. 가장 많이 나오는 단어 Top 10 를 returen 해주는 함수
#    예시 : top_ten_word(msg)

def top_ten_word(msg):
    top_ten = []  # 상위 10개의 단어를 저장할 리스트
    word_cnt = {}  # 각 단어의 빈도를 저장할 딕셔너리

    for w in msg.split():  # 입력된 메시지를 공백 기준으로 단어 분리
        nw = w.lower()  # 대소문자 구분 없이 처리하기 위해 소문자로 변환
        nw = nw.replace(".", "")  # 구두점 제거
        nw = nw.replace(",", "")
        nw = nw.replace('"', "")
        nw = nw.replace("'s", "")  # 소유격 표현 제거
        nw = nw.replace(":", "")
        nw = nw.replace(";", "")
        nw = nw.replace("?", "")
        nw = nw.replace("!", "")
        
        if not nw.isalpha():  # 숫자나 특수문자가 포함된 경우 제외
            # print(nw)  # 확인용 출력
            continue
        
        if len(nw) <= 3:  # 길이가 3 이하인 단어는 제외 (의미 없는 짧은 단어 제외 목적)
            continue
        
        word_cnt[nw] = word_cnt.get(nw, 0) + 1  # 단어 빈도 누적

    # print(word_cnt)  # 단어 빈도 딕셔너리 출력

    cnt_word = {}  # 빈도 수를 key로, 해당 빈도의 단어 리스트를 value로 저장하는 딕셔너리
    for w in word_cnt:
        cnt = word_cnt[w]  # 해당 단어의 빈도수
        cnt_word[cnt] = cnt_word.get(cnt, []) + [w]  # 동일 빈도의 단어 리스트에 추가

    # print(cnt_word)  # 빈도 기준 단어 리스트 출력

    for cnt in sorted(cnt_word.keys(), reverse=True):  # 빈도 수를 기준으로 내림차순 정렬
        if len(top_ten) <= 10:  # top_ten 리스트가 10개 이하일 경우만 추가
            top_ten.extend(cnt_word[cnt])  # 동일 빈도의 단어들 추가
        else:
            break  # 10개가 넘으면 중단

    return top_ten  # 최종적으로 상위 단어 리스트 반환

top_ten_word(msg)
```

self : 클래스 안 범위의 변수

```python
class Counter:
    def __init__(self): # 클래스 내의 변수 선언 및 초기화
        self.count = 0
    def reset(self):
        self.count = 0
    def increment(self):
        self.count += 1
    def get(self):
        return self.count
    

a = Counter()
# print(f"카운터 a의 값은 {a.get()}") # 오류발생
a.reset()
print(f"카운터 a의 값은 {a.get()}")
a.increment()
print(f"카운터 a의 값은 {a.get()}")
```

```python
class SuperCar(Car):
    def __init__(self, color, speed=0, bTurbo= True):
        super().__init__(color, speed)
        self.bTurbo = bTurbo
    
    def setTurbo(self, bTurbo =True):
        self.bTurbo = bTurbo

    def speedUp(self):
        if self.bTurbo:
            self.speed += 50
        else:
            super().speed += 50

    def __str__(self):
        if self.bTurbo:
            return f"[{self.color}] [speed = {self.speed}] 터보모드"
        else:   
            return f"[{self.color}] [speed = {self.speed}] 일반모드"
```

```python
# SuperCar 클래스는 Car 클래스를 상속받는다
class SuperCar(Car):
    # 생성자: 색상(color), 속도(speed, 기본값 0), 터보 여부(bTurbo, 기본값 True)를 초기화
    def __init__(self, color, price=0, speed=0, bTurbo=True):
        # 부모 클래스(Car)의 생성자를 호출하여 color와 speed를 초기화
        super().__init__(color, price, speed)
        # SuperCar만의 속성인 터보 모드를 초기화
        self.bTurbo = bTurbo

    # 터보 모드를 설정하는 메서드
    def setTurbo(self, bTurbo=True):
        self.bTurbo = bTurbo

    # 속도를 증가시키는 메서드
    def speedUp(self):
        # 터보 모드일 때는 속도를 50만큼 증가
        if self.bTurbo:
            self.speed += 50
        # 터보 모드가 아닐 때도 속도를 50만큼 증가
        else:
            super().speedUp()

    # 객체 정보를 문자열로 표현하는 메서드
    def __str__(self):
        # 터보 모드일 경우 출력 형태
        if self.bTurbo: # 터보모드가 True 일 때
            return f"[{self.color}] [speed = {self.speed}] 터보모드"
        # 일반 모드일 경우 출력 형태
        else:   
            return f"[{self.color}] [speed = {self.speed}] 일반모드"
```

```python
# 클래스가 클래스를 상속할 수 있음 - 상속성 (객체지향 프로그래밍 방법의 특징)

s1 = SuperCar('Gold', 0, 100, True)
s2 = SuperCar("White", 0, 0, False)

print(f"슈퍼카1: {s1}")
print(f"슈퍼카2: {s2}")
```

```python
class Person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age

    def setName(self, new_name):
        self.__name = new_name

    def setAge(self, new_age):
        self.__age = new_age

    def getName(self):
        return self.__name
    
    def getAge(self):
        return self.__age
    
    # 새로운 코드 추가
    def __eq__(self, personB): 
        return (self.__name == personB.__name) and (self.__age == personB.__age)

    def __str__(self): # print 가능하게 해줌
        return f"name:{self.__name}, age:{self.__age}"

a = Person("kim", 30)
b = Person("lee", 25)
print(a, b)
class Student(Person):
    def __init__(self, name, age, major):
        super().__init__(name, age)
        self.__major = major

    def setMajor(self, new_major):
        self.__major = new_major

    def getMajor(self):
        return self.__major
    
    def __eq__(self, personB): 
        return super().__eq__(personB) and (self.__major == personB.__major)
    
    def __str__(self): # self 재정의
        return f"name:{super().getName()}, age:{super().getAge()}, major:{self.__major}"

c = Student("ko", 30, "computer science")
d = Student("ku", 30, "computer science")

print(c)
c.setName("ku")
print(c)
print(d)
print("c == d:", c == d)
```
